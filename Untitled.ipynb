```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map( df )
```

```python
we can see on this graph the correlation between the columns. The value of the square of the same column are in diagonal and have the number 
one which means that the correlation is perfect, the squares with a value below 0 have a negative correlation, we can take the example of Pclass
and fare  or parch and age.
Being able to see the correlation between the columns allows us to see if certain factors are related to each others.
```
